### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Asumiendo que los DataFrames tienda1, tienda2, tienda3, tienda4 ya están cargados
# con los datos de cada tienda respectivamente.

# Calcular el ingreso total para cada tienda sumando la columna 'Precio'
ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

# Mostrar los resultados
print(f"Ingreso Total Tienda 1: ${ingreso_tienda1:,.2f}")
print(f"Ingreso Total Tienda 2: ${ingreso_tienda2:,.2f}")
print(f"Ingreso Total Tienda 3: ${ingreso_tienda3:,.2f}")
print(f"Ingreso Total Tienda 4: ${ingreso_tienda4:,.2f}")

# Opcionalmente, puedes guardar los resultados en un diccionario o DataFrame
ingresos_totales = {
    'Tienda 1': ingreso_tienda1,
    'Tienda 2': ingreso_tienda2,
    'Tienda 3': ingreso_tienda3,
    'Tienda 4': ingreso_tienda4
}
print("\nIngresos totales por tienda:")
print(ingresos_totales)
```

# 2. Ventas por categoría

```python
# --- Análisis para Tienda 1 ---
print("--- Ventas por Categoría - Tienda 1 ---")
# Agrupa por categoría, cuenta las ventas en cada una y ordena de mayor a menor
ventas_categoria_t1 = tienda.groupby('Categoría del Producto').size().sort_values(ascending=False)
# Muestra los resultados
print(ventas_categoria_t1)
# Muestra la categoría más popular
if not ventas_categoria_t1.empty:
    print(f"Categoría más popular Tienda 1: {ventas_categoria_t1.index[0]} ({ventas_categoria_t1.iloc[0]} ventas)")
print("\n" + "="*40 + "\n") # Separador

# --- Análisis para Tienda 2 ---
print("--- Ventas por Categoría - Tienda 2 ---")
ventas_categoria_t2 = tienda2.groupby('Categoría del Producto').size().sort_values(ascending=False)
print(ventas_categoria_t2)
if not ventas_categoria_t2.empty:
    print(f"Categoría más popular Tienda 2: {ventas_categoria_t2.index[0]} ({ventas_categoria_t2.iloc[0]} ventas)")
print("\n" + "="*40 + "\n") # Separador

# --- Análisis para Tienda 3 ---
print("--- Ventas por Categoría - Tienda 3 ---")
ventas_categoria_t3 = tienda3.groupby('Categoría del Producto').size().sort_values(ascending=False)
print(ventas_categoria_t3)
if not ventas_categoria_t3.empty:
    print(f"Categoría más popular Tienda 3: {ventas_categoria_t3.index[0]} ({ventas_categoria_t3.iloc[0]} ventas)")
print("\n" + "="*40 + "\n") # Separador

# --- Análisis para Tienda 4 ---
print("--- Ventas por Categoría - Tienda 4 ---")
ventas_categoria_t4 = tienda4.groupby('Categoría del Producto').size().sort_values(ascending=False)
print(ventas_categoria_t4)
if not ventas_categoria_t4.empty:
    print(f"Categoría más popular Tienda 4: {ventas_categoria_t4.index[0]} ({ventas_categoria_t4.iloc[0]} ventas)")
print("\n" + "="*40 + "\n") # Separador

# Opcional: Puedes guardar estos resultados en un diccionario si lo necesitas
# ventas_por_categoria = {
#     'Tienda 1': ventas_categoria_t1,
#     'Tienda 2': ventas_categoria_t2,
#     'Tienda 3': ventas_categoria_t3,
#     'Tienda 4': ventas_categoria_t4
# }
```

# 4. Productos más y menos vendidos

```python
import matplotlib.pyplot as plt

# Lista con las URLs de los archivos CSV de cada tienda
urls = [
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"
]

# Leemos cada archivo CSV y lo almacenamos en una lista llamada "tiendas"
# Cada elemento de la lista será un DataFrame que representa una tienda
tiendas = [pd.read_csv(url) for url in urls]

# Recorremos cada DataFrame (una tienda) junto con su índice (para numerarla)
for i, tienda in enumerate(tiendas, 1):

    # Contamos cuántas veces aparece cada producto (es decir, cuántas veces fue vendido)
    conteo = tienda['Producto'].value_counts()

    # Obtenemos el producto más vendido (el que más aparece)
    mas_vendido = conteo.idxmax()   # nombre del producto
    mas_vendido_cantidad = conteo.max()  # cantidad de ventas

    # Obtenemos el producto menos vendido (el que menos aparece)
    menos_vendido = conteo.idxmin()  # nombre del producto
    menos_vendido_cantidad = conteo.min()  # cantidad de ventas

    # Imprimimos los resultados en consola
    print(f"Tienda {i}:")
    print(f"  Producto más vendido: {mas_vendido} ({mas_vendido_cantidad} ventas)")
    print(f"  Producto menos vendido: {menos_vendido} ({menos_vendido_cantidad} ventas)")
    print("-" * 50)

    # Seleccionamos los 10 productos más vendidos para visualizarlos
    top10 = conteo.head(10)

    # Creamos un gráfico de barras horizontal
    top10.plot(kind='barh', figsize=(10, 6), color='skyblue', edgecolor='black')

    # Invertimos el orden para que el más vendido quede arriba
    plt.gca().invert_yaxis()

    # Agregamos título y etiquetas
    plt.title(f'Tienda {i} - Top 10 productos más vendidos')
    plt.xlabel('Cantidad de ventas')
    plt.ylabel('Producto')

    # Mostramos el gráfico
    plt.show()
```

# 5. Envío promedio por tienda

```python
# Recorremos cada tienda y calculamos el costo de envío promedio
for i, tienda in enumerate(tiendas, 1):
    # Obtenemos la media (promedio) de la columna 'Costo de envío'
    costo_envio_promedio = tienda['Costo de envío'].mean()

    # Mostramos el resultado con 2 decimales
    print(f"Tienda {i} - Costo de envío promedio: ${costo_envio_promedio:,.2f}")
```
